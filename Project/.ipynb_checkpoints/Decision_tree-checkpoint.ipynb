# Decision Tree

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Data_set/salaries.csv')
df.head(10)
```

```python
X=df.drop('salary_more_then_100k',axis='columns')
X
```

```python
Y=df['salary_more_then_100k']
Y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
X['le_company']=le_company.fit_transform(X['company'])
X['le_job']=le_job.fit_transform(X['job'])
X['le_degree']=le_degree.fit_transform(X['degree'])
```

```python
X
```

```python
x=X.drop(['company','job','degree'],axis='columns')
x
```

```python
from sklearn import tree

model=tree.DecisionTreeClassifier()
```

```python
model.fit(x,Y)
```

```python
model.predict([[0,0,0]])
```

```python
model.predict([[0,0,1]])
```
